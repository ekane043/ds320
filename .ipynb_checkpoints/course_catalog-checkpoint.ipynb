```python
y = soup.select("ul.childrenList a")
```

```python
for i in y:
    i = i.get('href','None')
```

```python
Course_Dict = {"major":[],
              "course_num":[],
              "course_title":[],
              "hours":[],
              "gened":[],
              "prereq":[],
              "course_description":[]}

for i in y:
    i = i.get('href','None')
    url = "https://www.luther.edu/catalog/curriculum/{}".format(i)
    requrl = requests.get(url)
    soup = BeautifulSoup(requrl.text,'html.parser')
    number = soup.find_all('span',class_ = "courseNumber")
    title = soup.find_all('span',class_ = "courseTitle")
    description = soup.find_all('span',class_ = "courseDescription")
    gened = soup.find_all('span',class_ = "genEd")
    for a in number:
        a = a.text
        Course_Dict['major'].append(i)
        Course_Dict['course_num'].append(a)
    for b in title:
        b = b.text
        Course_Dict['course_title'].append(b)
    for c in description:
        c = c.text
        Course_Dict['course_description'].append(c)
    for d in soup.select('ul.fa-ul'):
        d = d.text
        d = d[1]
        Course_Dict['hours'].append(d)
    fulfills = []
    fulfills_count = 0
    for e in soup.find_all('span',class_ = "genEd"):
        e = e.text
        fulfills.append(e)
    for x in soup.select('ul.fa-ul'):
            if "Fulfills:" not in x.text:
                Course_Dict['gened'].append("None") 
            else:
                Course_Dict['gened'].append(fulfills[fulfills_count])
                fulfills_count +=1
    for x in soup.select('ul.fa-ul'):
        if "Prerequisites:" in x.text:
            x = x.text
            x = x[23:]
            Course_Dict['prereq'].append(x)
        else:
            Course_Dict['prereq'].append('None')
```
